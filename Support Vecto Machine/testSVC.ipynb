```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from SVC import SoftMarginSVC,SoftMarginSVCPrimal,HardMarginSVC
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r"F:\mostafa\ML\course ML arabic Ahmed Rady\Day 37\nonlinear_data.csv")
data
```

```python
x = data.drop(["Label"],axis=1)
y = data["Label"] 
y = np.where(y==0,-1,1)
```

```python
scaler = StandardScaler()
X_Std = scaler.fit_transform(x)
```

```python
model = HardMarginSVC()
model.fit(X_Std,y)
plt.scatter(X_Std[:,0],X_Std[:,1],c = y)
xx1, xx2 = np.meshgrid(
    np.arange(-2,2,.01),
    np.arange(-2,2,.01)
)
meshgred_points = np.array([xx1.ravel(),xx2.ravel()]).T
z = model.predict(meshgred_points)
z = z.reshape(xx1.shape)
plt.contourf(xx1, xx2, z,alpha = .2)
print(model.score(X_Std,y))
plt.contourf(xx1,xx2,z,alpha=0.2)
```
